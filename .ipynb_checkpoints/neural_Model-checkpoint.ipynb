# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
X = df.drop("koi_disposition", axis=1)
y = df["koi_disposition"]
print(X.shape, y.shape)
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler


X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
y_train.unique()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)
```

```python
for label, original_class in zip(encoded_y_train, y_train):
    print('Original Class: ' + str(original_class))
    print('Encoded Label: ' + str(label))
    print('-' * 12)
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
# Step 2: Convert encoded labels to one-hot-encoding
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
```

# Train the Model



```python
# first, create a normal neural network with 40 inputs, 6 hidden nodes, and 3 outputs
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model_deep_neural = Sequential()
model_deep_neural.add(Dense(units=100, activation='relu', input_dim=40))
model_deep_neural.add(Dense(units=100, activation='relu'))
model_deep_neural.add(Dense(units=3, activation='softmax'))
```

```python
model_deep_neural.summary()
```

```python
# Compile the model
model_deep_neural.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Fit the model to the training data
model_deep_neural.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=100,
    shuffle=True,
    verbose=3
)
```

```python
model_loss, model_accuracy = model_deep_neural.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
model_loss, model_accuracy = model_deep_neural.evaluate(
    X_train_scaled, y_train_categorical, verbose=3)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
encoded_predictions = model_deep_neural.predict_classes(X_test_scaled)
prediction_labels = label_encoder.inverse_transform(encoded_predictions)
```

```python
print(f"Predicted classes: {prediction_labels}")
print(f"Actual Labels: {list(y_test[:5])}")
```

```python
# Calculate classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, prediction_labels,
                            target_names=["CANDIDATE", "CONFIRMED", "FALSE POSITIVE"]))
```
